# Series and DataFrame

pandas provides two data structures that shape data into a readable form:

1. Series
2. Data frame

- <b>Series</b> -A pandas series is a one-dimensional data structure that comprises of a key-value pair. It is similar to a python dictionary, except it provides more freedom to manipulate and edit the data.
In short a Series is a List of data, inside a DataFrame

To initialize a series, use pd.Series().


- <b>Dataframe</b> -A pandas dataframe is a two-dimensional data-structure that can be thought of as a spreadsheet. A dataframe can also be thought of as a combination of two or more series. 

To initialize a series, use pd.DataFrame()

Inside a DataFrame, we store several Dictionary by index value. 



- Each Dictionary referes to a single-people record.. so what if we want to store multiple records in a same-
dictionary,.. well there are many ways to do that, one is to create a list of all records inside that dictionary


```python
people ={
    'first': 'Ritam',
    'last': 'Arnold',
    'email': 'ritamadhya@gmail.com'
}

people ={
    'first': ['Ritam','Arnold'],
    'last': ['Arnold','Schwarzenegger'],
    'email': ['ritamadhya@gmail.com','as@gmail.com']
}

people ={
    'first':['Ritam','Arnold','louis'],
    'last' :['Arnold','Schwarzenegger','Philip'],
    'email': ['ritamadhya@gmail.com','as@gmail.com','lp@gmail.com']
}
```

```python
#Now if we want to check the 'email' column, we can access it by that key

people['email']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# This returns the Series of data in email

df['email']
```

```python
type(df['email'])
```

```python
# we can access multiple series at the same time

df[['first','last']]
```

```python
df.columns
```

To access <b>rows</b> in a DataFrame, we can use either loc or iloc
- <b>iloc</b> access a row by its 'integer' index number,
- <b>loc</b> access a row by its lable

```python
df.iloc[0]
```

```python
df.iloc[[0,1]]
```

```python
df.iloc[0,1]
```

```python
df.iloc[[0,1],2]
```

```python
df.loc[[1,2]]
```
